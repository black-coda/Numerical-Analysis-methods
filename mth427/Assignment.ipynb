### Plot the graphs of the class of polynomials from orders 1 to 20

```python
import matplotlib.pyplot as plt
import numpy as np


# Define the Chebyshev polynomials
def chebyshev_polynomial(x, n):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        Tn_2 = 1
        Tn_1 = x
        for i in range(2, n + 1):
            Tn = 2 * x * Tn_1 - Tn_2
            Tn_2 = Tn_1
            Tn_1 = Tn
        return Tn

# Generate x values
x = np.linspace(-1, 1, 500)

# Plot the Chebyshev polynomials T1 to T20
plt.figure(figsize=(14, 10))
for n in range(1, 21):
    y = chebyshev_polynomial(x, n)
    plt.plot(x, y, label=f'T{n}(x)')

plt.title('Chebyshev Polynomials of the First Kind (T1 to T20)')
plt.xlabel('x')
plt.ylabel('Tn(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Plot the graph of the class of polynomials from order 1-10 in the interval x  [ 1, 1]

```python
def monic_chebyshev_polynomial(x, n):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        Tn_2 = 1
        Tn_1 = x
        for i in range(2, n + 1):
            Tn = 2 * x * Tn_1 - Tn_2
            Tn_2 = Tn_1
            Tn_1 = Tn
        # Normalize to monic form by dividing by the leading coefficient (2^(n-1))
        return Tn / (2**(n-1))

# Generate x values
x = [i / 200.0 for i in range(-200, 201)]  # from -1 to 1

# Plot the monic Chebyshev polynomials T1 to T10
plt.figure(figsize=(14, 10))
for n in range(1, 11):
    y = [monic_chebyshev_polynomial(xi, n) for xi in x]
    plt.plot(x, y, label=f'T{n}(x)')

plt.title('Monic Chebyshev Polynomials of the First Kind (T1 to T10)')
plt.xlabel('x')
plt.ylabel('Tn(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Legendre polynomials of orders n=2 to n=10 and Plot the graphs in the interval x  [ 1, 1]

```python
import matplotlib.pyplot as plt

# Define the Legendre polynomials
def legendre_polynomial(x, n):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        Pn_2 = 1
        Pn_1 = x
        for i in range(2, n + 1):
            Pn = ((2 * (i - 1) + 1) * x * Pn_1 - (i - 1) * Pn_2) / i
            Pn_2 = Pn_1
            Pn_1 = Pn
        return Pn

# Generate x values
x = [i / 200.0 for i in range(-200, 201)]  # from -1 to 1

# Plot the Legendre polynomials P2 to P10
plt.figure(figsize=(14, 10))
for n in range(2, 11):
    y = [legendre_polynomial(xi, n) for xi in x]
    plt.plot(x, y, label=f'P{n}(x)')

plt.title('Legendre Polynomials (P2 to P10)')
plt.xlabel('x')
plt.ylabel('Pn(x)')
plt.legend()
plt.grid(True)
plt.show()
```
